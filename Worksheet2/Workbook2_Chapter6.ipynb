```python
#Exercise 1
text = 'Business Analytics'
text[4]  # This will get the 5th character in the string, which is "i"
text[2:4]  # This will get the characters from the 3rd to the 4th, resulting in "us"
text[:4]  # This will get the characters from the beginning to the 4th, resulting in "Bus"
text[4:]  # This will get the characters from the 5th to the last, resulting in "iness Analytics"
text[-2]  # This get the 2nd character from the end, which is "c"
text[-4:-2]  # This get the characters from the 4th to the 3rd from the end, resulting in "ti"
text[-2:]  # This get the characters from the 2nd from the end to the end, resulting in "cs"
text[:-2]  # This will get the characters from the beginning to the 2nd from the end, which is "Business Analyti"
```
