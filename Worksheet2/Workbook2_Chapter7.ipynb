```python
#CHALLENGE


filename = input("Enter the file name: ")

filepath = f'/content/drive/My Drive/Colab Notebooks/{filename}'
try:
    totalconfidence, count = 0, 0
    with open(filepath, 'r') as f:
        for line in f:
            if line.startswith("X-DSPAM-Confidence:"):
                confidence= line.split(":")[1].strip()
                try:
                    totalconfidence += float(confidence)
                    count += 1
                except ValueError:
                    continue

    if count > 0:
        print(f"Average Spam Confidence: {totalconfidence / count:.2f}")
    else:
        print("No 'X-DSPAM-Confidence:' lines found.")

except FileNotFoundError:
    print(f"{filename} does not exist.")


```
