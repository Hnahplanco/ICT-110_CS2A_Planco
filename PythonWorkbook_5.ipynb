```python
# A. Create a four dimensional array of random integers in the interval [0, 12]
# The array has a shape of (2, 3, 4, 5) as an example
import numpy as np


four_dim = np.random.randint(0, 13, (2, 3, 4, 5))
print("Four dimensional array:")
print(four_dim)

# Get the dimensions, shape, and size of the array
dimensions = four_dim.ndim
shape = four_dim.shape
size = four_dim.size

# The result is a 4D array with random integers between 0 and 12
# The dimensions, shape, and size provide information about the array structure
print("\nDimensions:", dimensions)
print("Shape:", shape)
print("Size:", size)

# B. Get the maximum number for each dimension
max_per_dimension = [np.max(four_dim, axis=i) for i in range(dimensions)]
print("\nMaximum number for each dimension:")
print(max_per_dimension)

# The result is an array containing the maximum values across each dimension

# C. Multiply each element in the array by 4
multiplied = four_dim * 4
print("\narray after multiplication by 4:")
print(multiplied)

# The result is a new 4D array where each element is the original value multiplied by 4

```
