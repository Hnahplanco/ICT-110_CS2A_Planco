```python
  #CHAPTER 4
  #Exercise 1
  def compute_pay(hours, rate):
    regular_hours = 40
    overtime_multiplier = 1.5

    if hours <= regular_hours:
        pay = hours * rate
    else:
        regular_pay = regular_hours * rate
        overtime_hours = hours - regular_hours
        overtime_pay = overtime_hours * overtime_multiplier * rate
        pay = regular_pay + overtime_pay

    return pay


hours_worked = float(input("Enter hours worked: "))
hourly_rate = float(input("Enter hourly rate: "))

total_pay = compute_pay(hours_worked, hourly_rate)
print("Total pay:", total_pay)
```

```python
#CHALLENGE
def computegrade(score):
    if 97 <= score <= 100:
        grade = '1.0'
    elif 93 <= score < 97:
        grade = '1.25'
    elif 89 <= score < 93:
        grade = '1.5'
    elif 85 <= score < 89:
        grade = '1.75'
    elif 81 <= score < 85:
        grade = '2.0'
    elif 77 <= score < 81:
        grade = '2.25'
    elif 74 <= score < 77:
        grade = '2.5'
    elif 0 <= score < 74:
        grade = '3.0'
    else:
        grade = 'Invalid Score'

    return grade

score = float(input("Enter score: "))
result_grade = computegrade(score)
print("Grade:", result_grade)
```

```python
#CHALLENEGE 2
max_number = None
min_number = None

while True:
    user_input = input("Enter a number (or 'done' to finish): ")

    if user_input.lower() == 'done':
        break

    try:
        number = float(user_input)


        if max_number is None or number > max_number:
            max_number = number

        if min_number is None or number < min_number:
            min_number = number

    except ValueError:
        print("Invalid input. Please enter a valid number or 'done'.")

if max_number is not None and min_number is not None:
    print("Maximum:", max_number)
    print("Minimum:", min_number)
else:
    print("No valid numbers entered.")
```
